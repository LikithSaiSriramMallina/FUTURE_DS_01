### Objective:
Analyze retail sales data to identify revenue trends, category performance, regional insights, and discount impact to support business decision-making.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month_Name'] = df['Order Date'].dt.month_name()
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
profit_margin = (total_profit / total_sales) * 100

total_sales, total_profit, profit_margin
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18,5))

ax[0].text(0.5, 0.5, f'${total_sales:,.0f}', fontsize=22, ha='center')
ax[0].set_title('Total Revenue')
ax[0].axis('off')

ax[1].text(0.5, 0.5, f'${total_profit:,.0f}', fontsize=22, ha='center')
ax[1].set_title('Total Profit')
ax[1].axis('off')

ax[2].text(0.5, 0.5, f'{profit_margin:.2f}%', fontsize=22, ha='center')
ax[2].set_title('Profit Margin')
ax[2].axis('off')

plt.tight_layout()
plt.show()
```

```python
monthly_sales = df.set_index('Order Date')['Sales'].resample('ME').sum()

plt.figure(figsize=(12,6))
monthly_sales.plot()
plt.title('Monthly Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()
```

```python
category_summary = df.groupby('Category')[['Sales','Profit']].sum()

category_summary.plot(kind='bar', figsize=(10,6))
plt.title('Category-wise Sales and Profit')
plt.xticks(rotation=0)
plt.show()
```

```python
region_summary = df.groupby('Region')[['Sales','Profit']].sum()

region_summary.plot(kind='bar', figsize=(10,6))
plt.title('Regional Sales and Profit')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit Relationship')
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top_products.plot(kind='barh')
plt.title('Top 10 Products by Sales')
plt.gca().invert_yaxis()
plt.show()
```

# Executive Summary

### Key Findings:
- Total Revenue: $2.29M
- Total Profit: $286K
- Profit Margin: 12.47%

### Category Insights:
- Technology is the most profitable category.
- Office Supplies provides stable margins.
- Furniture generates high revenue but low profit due to heavy discounting.

### Strategic Recommendations:
1. Cap Furniture discounts at 20%.
2. Focus on expanding Technology category.
3. Improve pricing strategy in low-performing regions.
4. Replace deep discounts with bundle promotions.
5. Prioritize high-margin products.
